```python
from keras.models import load_model
from keras.preprocessing.image import load_img, img_to_array
import os, shutil
import numpy as np
from skimage.transform import resize
#This program takes 2+ models and displays comparison of prediction float outputs on like images
#Variables to edit
imageheight=200
imagewidth=300
cutoff=0.7 #Above which is an echolcation and below is abnormal file
mfolder1path = "./weights/39ep40dropout4.h5" #Folder and model with saved model
ifolder = "./sorted/predict2/bad abnormals/" #Folder with images to test model with
mfoldercheck2 = True #Do you want to add the 2nd model to the comparison
mfolder2path = "./models/CNN200x300b.h5" #Folder and model with saved model
mfoldercheck3 = True #Do you want to add the 3rd model to the comparison
mfolder3path = "./weights1/383sgdep700postedonGithub.h5" #Folder and model with saved model

ifoldercheck2 = True #Do you want to check the 2nd folder of images?
ifolder2 = "./sorted/predict1/abn goodcalls/" #Folder with images to test model with
ifoldercheck3 = False #Do you want to check the 3rd folder of images?
ifolder3 = "../abcopy/a/" #Folder with images to test model with
#End of variables to edit

def ModelCompare(folder):
    model = load_model(mfolder1path)
    if mfoldercheck2 == True:
        model2 = load_model(mfolder2path)
    if mfoldercheck3 == True:
        model3 = load_model(mfolder3path)
    filelist = os.listdir(folder) #Make list of image filenames
    print('Processing folder with {0:2d} images at cutoff {1:2.1f}'.format(len(filelist),cutoff))
    print('\t\t\tModel1\t\tModel2\t\tModel3'.format(len(filelist)))
    for inum in range(len(filelist)):
        image = load_img(folder + filelist[inum], target_size=(imageheight,imagewidth))
        imageresize = image.resize((imagewidth, imageheight)).convert('RGB') #Resize image for desired CNN model
        imagearray = img_to_array(imageresize) #Make image into an array
        imagearray = np.expand_dims(imagearray, 0) #Add 4th dimension to array for CNN -> will get error otherwise
        imagearray = imagearray.astype('float32')/255 #Convert array so that its predictions are readable
        prediction=model.predict(imagearray) #Run prediction on image-Returns list of float predictions
        if mfoldercheck2 == True:
            prediction2=model2.predict(imagearray) #Run prediction on image-Returns list of float predictions
        else:
            prediction2=[0]
        if mfoldercheck3 == True:
            prediction3=model3.predict(imagearray) #Run prediction on image-Returns list of float predictions
        else:
            prediction3=[0]
        print('{0:s}\t{1:s}\t{2:s}\t{3:s}'.format(filelist[inum],str(prediction[0]),str(prediction2[0]),str(prediction3[0])))   

ModelCompare(ifolder)      
if ifoldercheck2 == True:
    ModelCompare(ifolder2)
if ifoldercheck3 == True:
    ModelCompare(ifolder3)        
print('Finished')    
```
